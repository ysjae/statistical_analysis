```python
def E(X,g=lambda x:x):
    x_set,f=X
    return np.sum([g(x_k)*f(x_k) for x_k in x_set])
E(X)
```

```python
mean=E(X)
```

```python
np.sum([(x_k-mean)**2*f(x_k) for x_k in x_set])
```

```python
mean=E(X)
np.sum([(x_k-mean)**2*f(x_k) for x_k in x_set])
```

```python
def V(X,g=lambda x: x):
    x_set, f= X
    mean=E(X,g)
    return np.sum([(g(x_k)-mean)**2*f(x_k) for x_k in x_set])
```

```python
V(X)
```

```python
V(X,lambda x:2*x+3)
```
